```python
from functions import wordle_filter, compare_words, score_fn, choose_opt
```

```python
import json

with open("wordle_clean_list.json", "r") as fp:
    wordle_list = json.load(fp)
```

```python
def solver(word_list, solution, init_word='cairo', verbose=False):
    in_lists = []
    not_l_lists = []
    in_n_lists = ['.', '.', '.', '.', '.']
    not_n_lists = [[], [], [], [], []]
    not_w_lists = []
    solved = False
    iter = 1
    remain_list = word_list

    while not solved:
        if verbose: print(iter)
        # choose optimal word
        if iter == 1:
            try_word = init_word
        else:
            try_word, best_sol_list = choose_opt(remain_list)
        if verbose: print(try_word)

        # check coincidencies    
        if try_word == solution:
            solved = True
            if verbose: print('solution:', try_word)
            return try_word, iter
        else:
            not_w_lists.append(try_word)
            in_lists, not_l_lists, in_n_lists, not_n_lists = compare_words(solution, try_word, in_list = in_lists, not_l_list = not_l_lists, in_n_list = in_n_lists, not_n_list = not_n_lists)
            if verbose: print(in_lists, not_l_lists, in_n_lists, not_n_lists, not_w_lists)

            # filt list
            remain_list = wordle_filter(remain_list, in_list=in_lists, not_l_list=not_l_lists, in_n_list=in_n_lists, not_n_list=not_n_lists, not_w_list=not_w_lists)
            if verbose: print(remain_list)
            iter += 1
```

```python
solver(wordle_list, 'anual')
```

```python
import statistics
total_tries = []
initial_word = 'cairo'

for word in wordle_list:
    solution, iter_need = solver(wordle_list, word, init_word=initial_word)
    if iter_need > 6:
        print(solution, iter_need)
    total_tries.append(iter_need)
average_tries = statistics.fmean(total_tries)
print(average_tries)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np

plt.hist(total_tries, bins=7, weights=np.ones(len(total_tries)) / len(total_tries))
plt.title('Number of attempts to solve all words')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.xlabel('Attempts ')
plt.ylabel('Total number of solutions')
```
